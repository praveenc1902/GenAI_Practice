```python
from transformers import BlipForImageTextRetrieval
```

```python
# Load model directly
from transformers import AutoProcessor, BlipForImageTextRetrieval

processor = AutoProcessor.from_pretrained("Salesforce/blip-itm-base-coco")
model = BlipForImageTextRetrieval.from_pretrained("Salesforce/blip-itm-base-coco")
```

```python
img_url = 'https://storage.googleapis.com/sfr-vision-language-research/BLIP/demo.jpg'
```

```python
from PIL import Image
import requests
```

```python
raw_image = Image.open(requests.get(img_url,stream=True).raw).convert('RGB')
```

```python
text1 = "image of a women and a dog sitting on the beach with tides coming and the women wore a white watch and dog has belt on dog body"
text2 = "image of a women and a dog sitting on the beach with tides"
```

```python
inputs = processor(images=raw_image,text=text1,return_tensors="pt")
```

```python
inputs
```

```python
itm_scores = model(**inputs)
```

```python
itm_scores
```

```python
itm_scores = model(**inputs)[0]
itm_scores
```

```python
import torch
itm_score = torch.nn.functional.softmax(
    itm_scores[0],dim=1)
print(f"""\
The image and text are matched \
with a probability of {itm_score[0][1]:.4f}""")
```

```python
inputs = processor(images=raw_image,text=text2,return_tensors="pt")
itm_scores = model(**inputs)[0]
itm_score = torch.nn.functional.softmax(
    itm_scores,dim=1)
print(f"""\
The image and text are matched \
with a probability of {itm_score[0][1]:.4f}""")
```

```python
inputs = processor(images=raw_image,text="there are 2 humans sitting on the hot coal mine",return_tensors="pt")
itm_scores = model(**inputs)[0]
itm_score = torch.nn.functional.softmax(
    itm_scores,dim=1)
print(f"""\
The image and text are matched \
with a probability of {itm_score[0][1]:.4f}""")
```
