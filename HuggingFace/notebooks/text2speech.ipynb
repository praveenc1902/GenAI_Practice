```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-to-speech", model="kakao-enterprise/vits-ljs")
```

```python
text = "Hey Praveen Reddy C, whats up ? how is everything going ?. Wish you happy Tuesday. Always be cheered uop buddy, love you lots"
```

```python
narr = pipe(text)
```

```python
narr
```

```python
from IPython.display import Audio as IPythonAudio

IPythonAudio(narr["audio"][0],
             rate=narr["sampling_rate"])
```
