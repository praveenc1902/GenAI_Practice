```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("budgetpdf.pdf")
data = loader.load()
```

```python
data
```

```python
data[0].page_content
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator="\n \n",
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len,
    is_separator_regex=False,
)
```

```python
docs_list = []
for i in range(0,len(data)):
    docs_list.append(data[0].page_content)
```

```python
docs_list[0]
```

```python

pdf_final_texts = text_splitter.create_documents(docs_list)
```

```python
len(pdf_final_texts)
```

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("unionbudgettext.txt")
textdata = loader.load()
```

```python
textdata
```

```python
from langchain_text_splitters import CharacterTextSplitter

text2_splitter = CharacterTextSplitter(
    separator="\\\n\\\n",
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len,
    is_separator_regex=False,
)
```

```python
text_final_texts = text2_splitter.create_documents([textdata[0].page_content])
```

```python
text_final_texts
```
